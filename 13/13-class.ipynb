## Cursors

This will help us return > 100 results for our searches!

```python
# cursor in twython -- cursors are relevant when you are "paging" through results, tracking your place
# .cursor() takes another function as a parameter
# then we put all the parameters we would have passed to our twitter.search function

cursor = twitter.cursor(twitter.search, q='"data journalism" -filter:retweets', count=100)
all_text = list()

for tweet in cursor:
    all_text.append(tweet['text'])
    if len(all_text) > 500: # twitter has rate limits, so we're building in a limit to our request 
        break # this aborts the loop
```

```python
all_text
```

```python
from collections import Counter
```

```python
# create a counter object
c = Counter()
```

```python
for item in all_text:
    c.update(item.split())

c.most_common(25)
```

**Stopwords** are words that are common across all texts, so you can safely exclude them (words like "the", "I", "you", etc.). Usually stopwords are shorter than other words. 

```python
c = Counter()
for item in all_text:
    c.update([w for w in item.split() if len(w) > 3])

c.most_common(25)
```

```python
c = Counter()
for item in all_text:
    # to fold both lowercase and uppercase versions as one word 
    c.update([w.lower() for w in item.split() if len(w) > 3])

c.most_common(25)
```

```python
cursor = twitter.cursor(twitter.search, q='"kevin durant" -filter:retweets', count=100)
all_urls = list()

for tweet in cursor:
    for item in tweet['entities']['urls']:
        all_urls.append(item['expanded_url'])
    if len(all_urls) > 1000: 
        break 
```

```python
all_urls
```

```python
url_count = Counter(all_urls)
for item in url_count.most_common(10):
    print(item)
```

```python
cursor = twitter.cursor(twitter.search, q='"data journalism" -filter:retweets', count=100)
all_media_urls = list()

for tweet in cursor:
    if 'media' in tweet['entities']:
        for item in tweet['entities']['media']:
            all_media_urls.append(item['media_url'])
    if len(all_media_urls) > 1000: 
        break 
```

```python
all_media_urls
```

```python
fh = open("preview.html", "w")
for item in all_media_urls:
    fh.write('<img src="{}" width="50">'.format(item))
fh.close()
```
